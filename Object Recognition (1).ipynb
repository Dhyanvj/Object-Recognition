```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf

from tensorflow import keras # keras is a neural network library that runs on TF
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
```

```python
# Load images
import glob
import os

# Define the path to the 'Photos' directory
photos_dir = r"C:\Users\dhyan\OneDrive - University of Hertfordshire\Documents\Object Recognition (1)\Training Photos"

# Get the list of image files in all subdirectories
image_files = glob.glob(os.path.join(photos_dir, '*/*.jpg'))

# Count the number of jpg files
image_count = len(image_files)

print(image_count)
```

```python
# Create a dataset
# Define some parameters for the loader:
batch_size = 32
img_height = 180
img_width = 180
```

```python
#  Split up my data into training and test sets:
train = tf.keras.utils.image_dataset_from_directory(
  photos_dir,
  validation_split = 0.2,
  subset = "training",
  seed = 123,
  image_size = (img_height, img_width),
  batch_size = batch_size)
```

```python
#  Split up my data into training and test sets:
test = tf.keras.utils.image_dataset_from_directory(
  photos_dir,
  validation_split = 0.2,
  subset = "validation",
  seed = 123,
  image_size = (img_height, img_width),
  batch_size = batch_size)
```

```python
class_names = train.class_names
print(class_names)
```

```python
# Visualize the data
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
for images, labels in train.take(1):
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off") 
```

```python
# Configure the dataset for performance

AUTOTUNE = tf.data.AUTOTUNE

train = train.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
test = test.cache().prefetch(buffer_size=AUTOTUNE)
```

```python
# Standardize the data
normalization_layer = layers.Rescaling(1./255)

normalized_ds = train.map(lambda x, y: (normalization_layer(x), y))
image_batch, labels_batch = next(iter(normalized_ds))
first_image = image_batch[0]
# Notice the pixel values are now in `[0,1]`.
print(np.min(first_image), np.max(first_image))
```

```python
# Greyscaling layer
greyscale_layer = layers.Lambda(lambda x: tf.image.rgb_to_grayscale(x))
```

```python
# Adjust contrast for histogram equalization
equalize_hist_layer = layers.Lambda(lambda x: tf.image.adjust_contrast(x, contrast_factor=2.0))
```

```python
# Combine all preprocessing layers
preprocessing_layers = Sequential([
    greyscale_layer,
    equalize_hist_layer,
    normalization_layer
])
```

```python
# Visualize preprocessed images
plt.figure(figsize=(10, 10))
for images, _ in train.take(1):
    for i in range(6):
        preprocessed_image = preprocessing_layers(images, training = True)
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(preprocessed_image[i].numpy().squeeze().astype("uint8"), cmap='gray')
        plt.axis("off")

plt.show()
```

```python
# Data augmentation
tf.get_logger().setLevel('ERROR')
data_augmentation = keras.Sequential(
  [
    layers.RandomFlip("horizontal",
                      input_shape=(img_height,
                                  img_width,
                                  3)),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
  ]
)
```

```python
plt.figure(figsize=(10, 10))
for images, _ in train.take(1):
    for i in range(9):
        augmented_images = data_augmentation(images, training=True)
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(augmented_images[0].numpy().astype("uint8"))
        plt.axis("off")

plt.show()
```

```python
# Using Dropout
num_classes = len(class_names)
model = Sequential([
  data_augmentation,
  preprocessing_layers,
  layers.Conv2D(16, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(32, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(64, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Dropout(0.2),
  layers.Flatten(),
  layers.Dense(128, activation='relu'),
  layers.Dense(num_classes, name="outputs")
])
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
# Model Summary
model.summary()
```

```python
# Train the model
epochs=30
history = model.fit(
  train,
  validation_data = test,
  epochs = epochs
)
```

```python
# Visualize training results
# Create plots of the loss and accuracy on the training and testing sets:

acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(epochs)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Testing Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Testing Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Testing Loss')
plt.legend(loc='upper right')
plt.title('Training and Testing Loss')
plt.show()
```

```python
# Path to the folder containing images
folder_path = r"C:\Users\dhyan\OneDrive - University of Hertfordshire\Documents\Object Recognition (1)\Photos (1)\Buildings_Valid"
# Get a list of all files in the folder
image_files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]

# Iterate over each image in the folder
for image_file in image_files:
    image_path = os.path.join(folder_path, image_file)

    # Loading and Preprocessing the Image:
    img = tf.keras.utils.load_img(
        image_path, target_size=(img_height, img_width)
    )
    img_array = tf.keras.utils.img_to_array(img)
    img_array = tf.expand_dims(img_array, 0) # Create a batch

    # Making Predictions with the Model:
    predictions = model.predict(img_array)
    score = tf.nn.softmax(predictions[0])

    # Displaying the Image and Prediction
    plt.imshow(img)
    plt.axis('off')
    plt.show()
    print(
        "This image most likely belongs to {} with a {:.2f} percent confidence."
        .format(class_names[np.argmax(score)], 100 * np.max(score))
    )
```
